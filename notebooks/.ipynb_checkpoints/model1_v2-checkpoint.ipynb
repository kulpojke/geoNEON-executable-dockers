```python
freq = '0.5H'

def data_to_pandas(df, cols, freq):
    '''changes the df from vaex to pandas'''
    # make timeBgn into datetime
    df['timeBgn'] = df.timeBgn.astype('datetime64')
    # create dask array of values from the df
    da = df[cols].to_dask_array()
    idx = df['timeBgn'].to_dask_array()
    # make a dask df
    ddf = dd.from_array(da, columns=cols)
    idx  = dd.from_array(idx)
    ddf = ddf.merge(idx.to_frame(name='timeBgn'))
        
    # convert to pandas df
    data = ddf.compute() 
    data.index = pd.to_datetime(data.timeBgn)
    data = data.asfreq(freq)
    #data['timeBgn'] = data.index
    
    return(data)

SITE = data_to_pandas(SITE, use_cols, freq)
```

## Data pre-processing

We will need to do something about nans and bad qf flags then turn columns into tensors for use in the model. __The bad qfs are all messed up at the moment so we will skip that part__

First we will fill all of the gaps shorter than some length threshold. For now lets say 5.

```python
def limited_gap_interpolate(df, n):
    '''Interpolates only gaps shorter than n'''
    mask = df.copy()
    for col in mask.columns:
        new_df = pd.DataFrame(df[col])
        new_df['new'] = ((new_df.notnull() != new_df.shift().notnull()).cumsum())
        new_df['ones'] = 1
        mask[col] = (new_df.groupby('new')['ones'].transform('count') < 5) | df[col].notnull()

    return(df.interpolate().bfill()[mask])

SITE = limited_gap_interpolate(SITE, 5)
```

Next let's drop columns that are just too NaN riddled to be redeemable... by use of an arbitray threshold. Also plot a heatmap of the NaNs ( not in function, just here for visualization.

```python
def prune_lame_columns(df, frac):
    '''Returns a version of df in which columns in which NaNs
    constitute a fraction greter than frac'''
    # find the columns that have more NaNs than the threshold
    na = df.isnull().sum()
    crappy_cols = list(na[na > len(df) * frac].index)

    # now make a list not containing the bad columns
    cols = [c for c in df.columns if c not in crappy_cols]

    # retrun df with only those good columns
    return(df[cols])


SITE = prune_lame_columns(SITE, 0.4)
# plot a heatmap
fig, ax = plt.subplots(figsize=(13,20));
sns.heatmap(SITE.isnull(), cbar=False, ax=ax);
```

Lets make a function to break the time series into chunks for each streak where not all values are NaN

```python
df = SITE
min_run_len = 600

def label_data_runs(df, min_run_len):
    # make a boolean df of NaN/not NAN 
    not_NaN = pd.DataFrame(df.isnull().all(1), columns=['nan'])
    not_NaN['notnan'] = ~not_NaN.nan

    # make a boolean column where True markse the start of a run of not-Nan
    not_NaN['start_run'] = not_NaN.notnan.ne(not_NaN.notnan.shift())

    # make an id for each non-NaN run (NaN streaks are labeled with id = 0 )
    not_NaN['run_id'] = not_NaN.notnan.astype('int') * not_NaN.start_run.cumsum()

    # count the run within each streak
    not_NaN['run_len'] = not_NaN.groupby('run_id').cumcount() + 1

    # make a df of the run_len, indexed by run_id
    runs = pd.DataFrame(not_NaN.groupby('run_id').run_len.max())
    runs = runs[runs.index != 0]
    runs['id'] = runs.index

    # make a dict full of dfs coresponding to each run
    dset_dict = {}
    for index, row in runs.iterrows():
        if row['run_len'] > min_run_len:
            idx = not_NaN[not_NaN.run_id == row['id']].index
            dset_dict[row['id']] = df.loc[idx]

    # for each df in dset_dict find a column with the short NaN runs and few NaNs
    
    # use tfp.sts.impute_missing_values (maybe) to impute missing values in that column
    
    
            
            
            
            
    return(dset_dict)        
            
d = label_data_runs(df, min_run_len)
len(d[3])
```
