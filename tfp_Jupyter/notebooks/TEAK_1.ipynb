```python
# clean up teak

# find all cols for SWC
swc_cols = [col for col in teak if 'VSWCMean' in col]

# find horizontal locs for SWC
hlocs = list(set([re.sub(r'^.*?h', 'h', col.partition('_')[2]) for col in swc_cols]))
hlocs.sort()

# create a horizontal average SWC for each depth
for h in hlocs:
    μcols = [col for col in teak.columns if (h in col) and ('VSWCMean' in col)]
    σcols = [col for col in teak.columns if (h in col) and ('VSWCVariance' in col)]

TEAK = teak[['secPrecipBulk', 'secPrecipExpUncert']]
TEAK['fluxCO2'] = teak['data.fluxCo2.nsae.flux']
TEAK['SWCμ'] = teak[μcols].mean(axis=1)
TEAK['SWCσ'] = teak[σcols].mean(axis=1)

# create a horizontal average ST for each depth
for h in hlocs:
    μcols = [col for col in teak.columns if (h in col) and ('soilTempMean' in col)]
    σcols = [col for col in teak.columns if (h in col) and ('soilTempVariance' in col)]

TEAK['STμ'] = teak[μcols].mean(axis=1)
TEAK['STσ'] = teak[σcols].mean(axis=1)

# drop NAs
b4 = len(TEAK)
TEAK = TEAK.dropna()
drpd = b4-len(TEAK)
print(f'dropped {drpd} out of {b4} entries ({round(100*drpd/b4, 2)} %) due to NAs')
```

```python


# find all cols for each data param
swc_cols = [col for col in df if 'VSWCMean' in col]
co2_cols = [col for col in df if 'soilCO2concentrationMean' in col]
st_cols =  [col for col in df if 'soilTempMean' in col]

# find locs for each data param
swc_locs = set([col.partition('_')[2] for col in swc_cols])
co2_locs = set([col.partition('_')[2] for col in co2_cols])
st_locs = set([col.partition('_')[2] for col in st_cols])

# find locs common to all
locations = list(swc_locs.intersection(co2_locs, st_locs))

# split up df in || 
@delayed
def splitter(df, loc):
    swc = 'VSWCMean_' + loc
    st  = 'soilTempMean_' + loc
    c   = 'soilCO2concentrationMean_' + loc
    d = df[[c, swc, st]]
    sensor = [loc] * len(d)
    d = d.assign(sensor=sensor)
    d.columns = ['SCO2', 'SWC', 'ST', 'sensor']
    return(d)

delayed_dfs = []
for loc in locations:
    delayed_dfs.append(splitter(df, loc))

# concat the dfs into a long df
new_df = pd.concat(compute(*delayed_dfs))
new_df = new_df.dropna()

data = new_df.reset_index()[['SCO2', 'SWC', 'ST']]
print(data.head())
```
